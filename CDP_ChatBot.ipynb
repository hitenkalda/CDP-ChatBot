```python
from sentence_transformers import SentenceTransformer, util
import json
import torch

# Load the extracted data
def load_data(filepath="cdp_documentation.json"):
    with open(filepath, "r", encoding="utf-8") as f:
        return json.load(f)

# Prepare the dataset for embeddings
def prepare_dataset(data):
    dataset = []
    for source, content in data.items():
        sentences = content.split(". ")  # Split into sentences for better granularity
        for sentence in sentences:
            dataset.append({"source": source, "content": sentence.strip()})
    return dataset

# Embed the dataset
def embed_dataset(dataset, model):
    for entry in dataset:
        entry["embedding"] = model.encode(entry["content"], convert_to_tensor=True)
    return dataset

# Answer user questions
def answer_question(question, dataset, model, threshold=0.5):
    question_embedding = model.encode(question, convert_to_tensor=True)
    similarities = []

    # Calculate similarities between the question and dataset entries
    for entry in dataset:
        similarity = util.pytorch_cos_sim(question_embedding, entry["embedding"])
        similarities.append((similarity.item(), entry))

    # Find the best match
    best_match = max(similarities, key=lambda x: x[0])
    best_score, best_entry = best_match

    if best_score < threshold:
        return f"⚠️ Warning: Your question seems out of scope. No relevant information found in the documentation.", None

    return f"Answer from {best_entry['source']}: {best_entry['content']}", best_score

if __name__ == "__main__":
    # Step 1: Load extracted documentation data
    data_filepath = "cdp_documentation.json"  # Ensure this file exists in the same directory
    documentation_data = load_data(data_filepath)

    # Step 2: Prepare and embed dataset
    model_name = "sentence-transformers/all-MiniLM-L6-v2"  # Lightweight, fast sentence transformer model
    model = SentenceTransformer(model_name)
    dataset = prepare_dataset(documentation_data)
    embedded_dataset = embed_dataset(dataset, model)

    # Step 3: Interactive Q&A Loop
    print("\nSupport Agent Chatbot is ready to answer your questions about CDPs!")
    print("Type 'exit' to end the session.\n")

    while True:
        user_question = input("Ask a 'how-to' question: ").strip()
        if user_question.lower() == "exit":
            print("Goodbye!")
            break

        answer, confidence = answer_question(user_question, embedded_dataset, model)
        print(f"\n{answer}\n")
        if confidence:
            print(f"Confidence: {confidence:.2f}\n")
```

```python
# Install required libraries
!pip install -q sentence-transformers beautifulsoup4 requests

from sentence_transformers import SentenceTransformer, util
from bs4 import BeautifulSoup
import requests
import json
import re
import time
from google.colab import files

# ------------- Web Scraping Section ------------- #

# Documentation URLs
DOC_URLS = {
    # "Segment": "https://segment.com/docs/",
    "mParticle Home": "https://docs.mparticle.com/",
    "mParticle Integrations": "https://docs.mparticle.com/integrations/",
    "mParticle Android SDK": "https://docs.mparticle.com/developers/client-sdks/android/initialization/",
    "Lytics Home": "https://docs.lytics.com/",
    "Lytics Quickstart": "https://docs.lytics.com/docs/developer-quickstart",
    "Zeotap Home": "https://docs.zeotap.com/home/en-us/",
    "Zeotap Release Notes": "https://docs.zeotap.com/articles/#!product-release-notes/1-6-2"
}

visited_urls = set()

# Fetch and parse HTML content
def fetch_html_content(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None

# Extract and clean text from HTML
def extract_clean_text(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    text = soup.get_text(separator="\n", strip=True)
    cleaned_text = re.sub(r"\s+", " ", text)  # Remove excessive whitespace
    cleaned_text = re.sub(r"\[.*?\]", "", cleaned_text)  # Remove bracketed content
    return cleaned_text

# Recursive crawling function
def crawl(url, max_depth=2, current_depth=0, delay=1):
    if current_depth > max_depth or url in visited_urls:
        return ""

    print(f"Crawling: {url}")
    visited_urls.add(url)
    content = ""

    html_content = fetch_html_content(url)
    if not html_content:
        return content

    soup = BeautifulSoup(html_content, "html.parser")
    content += extract_clean_text(html_content) + "\n\n"

    for link in soup.find_all("a", href=True):
        next_link = requests.compat.urljoin(url, link['href'])
        if next_link not in visited_urls and url in next_link:
            content += crawl(next_link, max_depth, current_depth + 1, delay)
            time.sleep(delay)  # Be polite

    return content

# Process all documentation URLs
def process_documentation(doc_urls):
    data = {}
    for name, url in doc_urls.items():
        print(f"Processing: {name}")
        data[name] = crawl(url, max_depth=2)
    return data

# Save scraped data to JSON
def save_to_json(data, filename="cdp_documentation.json"):
    filepath = f"/content/{filename}"
    with open(filepath, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print(f"Data saved to {filepath}")
    return filepath

# Download the JSON file in Colab
def download_file(filepath):
    print(f"Downloading {filepath}...")
    files.download(filepath)

# ------------- Chatbot Section ------------- #

# Load the scraped documentation data
def load_data(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        return json.load(f)

# Prepare data by splitting it into chunks
def prepare_dataset(data):
    dataset = []
    for source, content in data.items():
        sentences = re.split(r'[.?!]\s+', content)
        for sentence in sentences:
            if sentence.strip():
                dataset.append({"source": source, "content": sentence.strip()})
    return dataset

# Embed data using Sentence Transformer
def embed_dataset(dataset, model):
    for entry in dataset:
        entry["embedding"] = model.encode(entry["content"], convert_to_tensor=True)
    return dataset

# Answer user questions
def answer_question(question, dataset, model, threshold=0.5):
    question_embedding = model.encode(question, convert_to_tensor=True)
    similarities = []

    for entry in dataset:
        similarity = util.pytorch_cos_sim(question_embedding, entry["embedding"])
        similarities.append((similarity.item(), entry))

    best_match = max(similarities, key=lambda x: x[0])
    best_score, best_entry = best_match

    if best_score < threshold:
        return "⚠️ Warning: Your question seems out of scope. No relevant information found.", None

    return f"Answer from {best_entry['source']}: {best_entry['content']}", best_score

# ------------- Main Execution ------------- #

# Step 1: Scrape and save data
documentation_data = process_documentation(DOC_URLS)
output_file = save_to_json(documentation_data)
download_file(output_file)

# Step 2: Load data and train model
loaded_data = load_data(output_file)
model = SentenceTransformer("all-MiniLM-L6-v2")
dataset = prepare_dataset(loaded_data)
embedded_dataset = embed_dataset(dataset, model)

# Step 3: Interactive Chatbot
print("\n📘 Chatbot is ready! Ask me any 'how-to' question about Segment, mParticle, Lytics, or Zeotap.")
print("Type 'exit' to end the chat.\n")

while True:
    user_question = input("❓ Your Question: ").strip()
    if user_question.lower() == "exit":
        print("👋 Goodbye!")
        break

    answer, confidence = answer_question(user_question, embedded_dataset, model)
    print(f"\n{answer}\n")
    if confidence:
        print(f"🔎 Confidence Score: {confidence:.2f}\n")
```
